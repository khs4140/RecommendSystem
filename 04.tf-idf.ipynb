```python
removal_list =  "‘, ’, ◇, ‘, ”,  ’, ', ·, \“, ·, △, ●,  , ■, (, ), \", >>, `, /, -,∼,=,ㆍ<,>, .,?, !,【,】, …, ◆,%"


# 특수기호 전처리 함수
def cleansing_special(sentence):
    sentence = re.sub('[.,\'\"''""~?]', '', sentence)
    sentence = re.sub('[^가-힣0-9a-zA-Z\\s]', ' ', sentence)
    sentence = re.sub("\s+", " ", sentence)

    sentence = sentence.translate(str.maketrans(removal_list, ' '*len(removal_list)))
    sentence = sentence.strip()

    return sentence

# 숫자 전처리 (delexicalization) 함수
def cleansing_number(sentence):
    sentence = re.sub('[0-9]+', 'NUM', sentence)
    sentence = re.sub('NUM\s+', 'NUM', sentence)
    sentence = re.sub('[NUM]+', 'NUM', sentence)

    return sentence
```

```python
sentences_list = []
with open('./data/news.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    for i, line in enumerate(lines):
        pre_line = cleansing_special(line)
        pre_line = cleansing_number(line)

        data = ' '.join(list(set(mecab.nouns(pre_line))))

        if (i+1) % 5 == 0:
            print(f'=={i+1}번 문장 ==')
            print(data)
        sentences_list.append(data)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_mat = tfidf.fit_transform(sentences_list)

print(tfidf_mat.shape)
```

```python
from collections import defaultdict

word2id = defaultdict(lambda : 0)
for idx, feature in enumerate(tfidf.get_feature_names()):
    word2id[feature] = idx
```

```python
for idx, line in enumerate(sentences_list):
    
    results = [(token, tfidf_mat[idx, word2id[token]]) for token in line.split()]
    results.sort(key = lambda x: x[1], reverse=True)
    
    if (idx+1) % 5 == 0:
        print(f'--- 문서 {idx+1}번 ---')
        print(results)
        print('\n')
```
